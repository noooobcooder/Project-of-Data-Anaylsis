***EDA + Graphs***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load clean dataset
df = pd.read_csv('/content/netflix_cleaned.csv')
```

```python
df.head()
```

***Basic EDA (Summary and info)***

```python
# Shape of Dataset
(df.shape)
```

```python
# Columns
(df.columns)
```

```python
# Data types
(df.dtypes)
```

```python
# Missing value (it should be zero because we clean it alreday)
(df.isnull().sum())
```

```python
# Movies vs TV shows
(df['type'].value_counts())
```

```python
# Top 10 countries with most content
(df['country'].value_counts().head(10))
```

```python
# Content Release by year(last 15 year)
(df['release_year'].value_counts().sort_index().tail(15))
```

***GRAPHS (Visualization)***

```python
# Movies Vs TV Show
sns.countplot(x="type", data=df, palette="Set2")
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

```python
# Release Trend(BY Year)
df['release_year'].value_counts().sort_index().plot(kind='line', figsize=(10,5))
plt.title("Netflix Content Release Over Years")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
# Top 10 contries by Content
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="coolwarm")
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```

```python
# Title Length Distribution
sns.histplot(df['title'].apply(len), bins=20, kde=True, color="purple")
plt.title("Distribution of Title Lengths")
plt.xlabel("Title Length")
plt.show()
```
